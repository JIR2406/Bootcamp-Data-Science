````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sample_means = []
dado = pd.Series([1, 2, 3, 4, 5, 6])
for i in range(10):
    samp_5 = dado.sample(5, replace=True)
    sample_means.append(np.mean(samp_5))
plt.hist(sample_means)
plt.show()
````

La distribucion especificamente es una distribucion de muestreo de la media de la muestra.
- La distribucion de muestreo se aproximara a una distribucion normal a medida que aumenta el numero de ensayos.


El teorema del limite central o TLC se aplica tambien a otras estaisticas como la mediana, la varianza, etc.
