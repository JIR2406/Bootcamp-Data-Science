# Correlación
Lo primero que haremos sera hablar sobre las relaciones entre variables numericas.
Podemos visualizar este tipo de relaciones con graficos de dispersion.


La variable en el eje `x` se llama variable independiente/explicativa y la variable en el eje `y` se llema variable de respuesta/dependiente.

Tambien podemos examinar las relaciones entre dos variables numericas utilizando un numero llamado coeficiente de correlación.

### Coeficiente de correlación

Cuantifica la relacion lineal entre dos variables.
Se trata de un numero entre -1 y 1, donde la magnitud corresponde a la fuerza de la relacion.

- Cuantifica la relacion lineal entre dos variables.
- Numero entre -1 y 1.
- La magnitud corresponde a la fuerza de la relacion.
- El signo (+ o -) corresponde a la dirección de la relación.

Un coeficiente de correlacion alrededor de 0.2 se considera una relacion debil.
El signo del coeficiente de correlacion corresponde a la direccion de la relacion.

> Un coeficiente de correlación positivo indica que a medida que x aumenta, y tambien aumenta.

Utilizaremos seaborn como sns que es el alias comunmente utilizado para seaborn.


```python
import seaborn as sns
import pandas as pd

msleep = pd.read_csv('../../data/TMDB_movie_dataset_v11.csv')
msleep = msleep.head(100)
sns.scatterplot(x='vote_average',y='vote_count',data=msleep)
```

### Añadir una linea de tendencia
Si queremos la correlacion entre dos variables, podemos añadir una linea de tendencia a nuestro grafico de dispersion.

```python
import matplotlib.pyplot as plt
sns.lmplot(x='vote_average', y='vote_count', data=msleep, ci=None)
plt.show()
```

Tenga en cuenta que no importa en que serie se invoca el metodo y cual se pasa ya que la correlacion entre x e y es la misma que la correlacion entre y e x.


Hay mas de una forma de calcular la correlación, pero el metodo que hemos estado usando se llama correlacion producto-momento de Pearson, que tambien se escribe como r.
Esta es la medida de correlacion mas comunmente utilizada.

```python
msleep['vote_average'].corr(msleep['vote_count'])
```

Matematicamente se calcula utilizando esta forma donde las barras x e y son las medias de x e y respectivamente.

```python
msleep['vote_count'].corr(msleep['vote_average'])
```
