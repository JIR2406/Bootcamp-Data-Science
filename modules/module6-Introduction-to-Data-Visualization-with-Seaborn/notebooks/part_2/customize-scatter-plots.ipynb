# Descripción general del grafico de dispersión

- Subgraficos (col y row)
- Subgrupos con color (hue)
Nuevas personalizaciones:
- Subgrupos con tamaño y estilo de punto
- Cambiar la transparencia de los puntos

>El tamaño se basa en la cantidad de personas en el grupo, y los grupos más grandes tienen puntos más grandes en la trama.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            size='size')
plt.show()
```

Este grafico es un poco dificil de leer porque todos los puntos son del mismo color.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            size='size',
            hue='size')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',
            data=tips,
            kind='scatter',
            hue='smoker',
            style='smoker')
plt.show()
```

La ultima personalización que veremos es la transparencia de los puntos.
Establecemos el parametro `alpha` en 0.5 para que los puntos sean semitransparentes.
Esto es útil cuando hay muchos puntos superpuestos, ya que permite ver la densidad de los puntos en áreas donde se superponen.

```python
sns.relplot(x='total_bill',y='tip',
            data=tips,
            kind='scatter',
            hue='smoker',
            style='smoker',
            alpha=0.5)
plt.show()
```
