# ¿Qué son los gráficos de lineas?
En `Seaborn`, tenemos dos tipos de diagramas relacionales: diagramas de dispersion y diagramas de lineas.

- Gráficos de dispersion
Cada punto del gráfico es una observación independiente.
- Gráficos de lineas
Cada punto del gráfico representa lo mismo, normalmente registrado a lo largo del tiempo.


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.pyplot import errorbar

air_df_mean = pd.read_csv('../../data/countries-of-the-world.csv')
air_df_mean['']
air_df_mean.info()
````

````python
sns.relplot(x='',
            y='',
            data = air_df_mean,
            kind='scatter')
plt.show()
````

Al establecer el parametro `marcadores` en `True` se mostrara un marcador para cada punto de datos.

El marcador varia segun el subgrupo que haya configurado utilizando el parametro `style`.
Si no desea que los estilos de linea varien segun el subgrupo establezca el parametro guion='line' en el constructor de la figura.

# Varias observaciones por valor x
```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.relplot(x='year',
            y='lifeExp',
            data = air_df_mean,
            kind='scatter',
            markers=True)
plt.show()
```
Si a un grafico de lineas se le dan multiples observaciones por valor de x, las agregara en una unica medida de resumen.
Observese que Seaborn calculara automaticamente un intervalo de confianza para la media, que se muestra en la region sombreada.

La zona sombreada es el intervalo de confianza
- Supone que el conjnto de datos es una muestra aleatoria
- 95 % de confianza en que la media se encuentra dentro de este intervalo
- Indica incertidumbre en la estimacion de la media
- No es un intervalo de confianza para los datos individuales
- No es un intervalo de confianza para la media de la poblacion
