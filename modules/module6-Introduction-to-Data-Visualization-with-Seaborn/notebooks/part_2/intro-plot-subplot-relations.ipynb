Muchas preguntas en la ciencia de datos se centran en describir la relacion entre dos variables cuantitativas.
Seaborn llama a las tramas que visualizan esta relacion **tramas relacionales** entre dos variables cuantitativas y las respondimos con graficos de dispersion.

# Crear una grafica separada por subgrupos

`relplot()` significa grafico relacional y le permite visualizar la relacion entre dos variables cuantitativas utilizando graficos de dispersion o graficos de lineas.
- Crea graficos relacionales: Graficos de dispersion o graficos de lineas.
Â¿Por que utilizar `relplot()` en lugar de `scatterplot()`?
- `relplot()` te permite crear sibgraficos en una sola figura

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            col='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            col='day',
            col_wrap=2,
            col_order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.show()
```
