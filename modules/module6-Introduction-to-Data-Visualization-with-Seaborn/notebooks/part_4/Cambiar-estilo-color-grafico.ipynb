# ¿Por qué personalizar?
De forma predeterminada, los graficos de Seaborn son agradables a la vista, pero hay varias razones por las que es posible que desee cambiar su apariencia.
Cambiar el estilo de una trama puede estar motivado por la preferencia personal pero tambien puede ayudar a mejorar su legibilidad o ayudar a orientar ala audiencia mas rapidamente hacia la conclusión clave.

- Preferencia personal
- Mejorar la legibilidad
- Interpretación de guías

# Cambiar el estilo de las figuras
Seaborn tiene cinco estilos de figuras preestablecidos que cambian el fondo y los ejes de la trama.
- La figura estilo incluye el fondo y los ejes de la trama.
- Opciones preestablecidas blanco, oscuro, cuadricula blanca, cuadricula oscura, marcas.

Para establecer uno de estos como estilo global para todos sus graficos, utilice la funcion `establecer_estilo()`.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
bins = sns.load_dataset("penguins")
sns.set_style("whitegrid")
sns.set_palette(['#39A7D0','#36ADA4'])
sns.set_context("notebook")
sns.catplot(x='sex',
            y='body_mass_g',
            hue='sex',
            data=bins,
            kind='box',
            estimator=median)
plt.show()
```

# Cambiar la paleta
- La figura `paleta` cambia el color de los elementos principales del gráfico
- `sns.set_palette()`
- Utiliza paletas predefinidas o crea una paleta personalizada.

La visualización trata con una escala donde los dos extremos de la escala son opuestos y hay un punto medio neutral.

Otros grupos de paletas se denominan paletas secuenciales, se trata de un unico color que van desde valores claros a oscuros.
- Paletas secuenciales: `Blues`, `Greens`, `Purples`, `Oranges`, `Reds`, `Greys`
- Paletas secuenciales divergentes: `coolwarm`, `RdBu`, `PiYG`, `PRGn`, `BrBG`, `PuOr`, `RdYlBu`, `RdYlGn`, `Spectral`
- Paletas cualitativas: `Set1`, `Set2`, `Set3`, `Pastel1`, `Pastel2`, `Dark2`, `Accent`, `Paired`, `tab10`, `tab20`, `tab20b`, `tab20c`

# Cambiar la escala
- La figura contexto cambia la escala de los elementos y las etiquetas del grafico.
- sns.set_context()
- De menor a mayor: papel, cuaderno, charla, cartel.
