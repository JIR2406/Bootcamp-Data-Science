```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos
df = pd.read_csv('../../data/student-alcohol-consumption.csv')
print("Datos cargados correctamente:")
print(df.columns)
```

```python
ax = sns.countplot(data=df, x="age", hue="age", palette='Set1', legend=False) # Usar 'hue' para colorear por edad
ax.set_facecolor("white")
ax.set_xlabel('Edad')
ax.set_ylabel('NÃºmero de estudiantes')
plt.show()
```
