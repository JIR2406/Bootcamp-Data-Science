# Conjunto de datos de propinas

Para mostrar esta caracteristica interesante de Seaborn, usaremos el conjunto de datos de sugerencias integrado de Seaborn.
Podemos acceder a el utilizando la funcionar `load_dataset()` de Seaborn.
Pasamos como argumento el nombre del conjunto de datos, que en este caso es "tips".

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head()
```

>Exploraremos la relación entre las columnas `total_bill`, `tip` usando un diagrama de dispersión.

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue = 'smoker',hue_order=['Yes','No'])
plt.show()
```

Es posible controlar los colores asignados a cada valor usando el argumento `palette`.
Este parametro incluye un diccionario, que es una estructura de datos que tiene pares clave-valor.
Este diccionario debe asignar los valores de las variables a los colores que desea representar el valor.

```python
hue_colors = {"Yes": "black",
              "No": "cyan"}
sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue = 'smoker',
                palette=hue_colors)
plt.show()
```

En el siguiente grafico podemos ver que los hombres superan en numero a las mujeres tanto entre los fumadores como entre los no fumadores en este conjunto de datos.

```python
sns.countplot(x="smoker",
              data=tips,
              hue='sex')
plt.show()
```
