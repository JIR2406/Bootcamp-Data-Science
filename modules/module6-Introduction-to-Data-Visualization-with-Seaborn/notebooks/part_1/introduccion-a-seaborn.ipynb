# ¿Qué es Seaborn?
Seaborn es una poderosa biblioteca de Python para crear visualizaciones de datos.
Fue desarrollada para facilitar la creación de los tipos de tramas mas comunes.
La siguiente imagen es un flujo tipico de analisis de datos.
<img src="../../resource/flujo-tipico.png">

La visualización de datos es a menudo un componente enorme tanto de la exploración de datos fase y la comunicación de resultados, por lo que Seaborn sera muy util allí.

# Ventajas de Seaborn
- Facilitar la visualización de datos.
- Funciona muy bien con estructuras de datos de Pandas.
- Esta construida sobre Matplotlib, por lo que puedes usar las funciones de Matplotlib junto con Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

height = [1.5, 1.6, 1.7, 1.8, 1.9, 2.0]
weight = [50, 60, 70, 80, 90, 100]
sns.scatterplot(x=height, y=weight)
plt.show()
```

# Gráficos de conteo
Los gráficos de conteo incluyen una lista categórica y devuelven barras que representan el número de entradas de la lista por categoría.

```python
gender = ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']
sns.countplot(x=gender)
plt.show()
```
