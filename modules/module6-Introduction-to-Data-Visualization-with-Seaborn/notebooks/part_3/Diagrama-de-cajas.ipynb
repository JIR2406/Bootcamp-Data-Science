# Crear un diagrama de cajas
- Muestra la distribución de datos cuantitativos.
- El cuadro de color representa el percentil 25 (Q1) y el percentil 75 (Q3).
- La linea en el medio del cuadro represente la mediana.
- Los "bigotes" dan una idea de la extensión de la distribución
- Los puntos flotantes representan valores atípicos (outliers).

Los diagramas de caja se utilizan comunmente como una forma de comparar la distribución de una variable cuantitativa entre diferentes grupos de una variable categórica.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('../../data/young-people-survey-responses.csv')
df.head(2)
```

```python
sns.catplot(x='Village - town',kind='box',y='Age',data=df,hue='Village - town')
plt.show()
```

> Podemos evitar los valores atípicos (outliers) usando el parámetro `sym=""`, en versiones mas recientes se usa `showfliers`.

```python
sns.catplot(x='Village - town',
            kind='box',
            y='Age',
            data=df,
            hue='Village - town',
            showfliers= False)
plt.show()
```

# Cambiar los bigotes con `whis`
- De forma predeterminada, los bigotes se extienden hasta 1.5 veces el rango intercuartilico.
- Haz que se extienda a 2,0 * IQR (rango intercuartilico) con el parámetro `whis=2.0`.
- Mostrar los perceptiles 5 y 95 con `whis=[5, 95]`.
- Mostrar valores minimos y maximos con `whis=[0, 100]`.

```python
g= sns.catplot(x='Village - town',
            kind='box',
            y='Age',
            data=df,
            hue='Village - town',
            whis=[0,100])
plt.show()
```
