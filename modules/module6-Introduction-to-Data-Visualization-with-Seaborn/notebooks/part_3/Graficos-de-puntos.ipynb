# ¿Qué son los gráficos de puntos?
Los graficos de puntos muestran la media de una variable cuantitativa para las observaciones de cada categoria trazada como un solo punto.

- Los puntos muestran la media de la variable cuantitativa
- Las lineas verticales muestran los intervalos de confianza del 95%

Tanto los graficos de lineas como los de puntos muestran la media de una variable cuantitativa y los intervalos de confianza del 95% para la media.

> Los graficos de lineas son graficos relacionales por lo que tanto el eje x como el eje y son variables cuantitativas.

> En un grafico de puntos, un eje (normalmente el eje x) es una variable categorica lo que lo convierte en un grafico categorico.


# Gráficos de puntos frente a gráficos de barras
En el grafico de puntos, es mas facil comparar las alturas de los puntos del subgrupo cuando estan apilados uno encima del otro.

```python
import seaborn as sns
import matplotlib.pyplot as plt

bins = sns.load_dataset("diamonds")
sns.catplot(x="cut",
            y="price",
            data=bins,
            kind="point",
            hue = "color")
plt.show()
```

A veces es posible que queramos eliminar las lineas que conectan cada punto, tal vez porque solo deseamos comparar dentro de un grupo de categorias y no entre ellos.

Para hacer esto establezca el parametro `join` a `False`:

Para hacer que los puntos y los intervalos de confianza se calculen para la mediana en lugar de la media importe la función mediana de la biblioteca numpy y establezca el estimador igual a la funcion mediana numpy.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
bins = sns.load_dataset("diamonds")
sns.catplot(x="cut",
            y="price",
            data=bins,
            kind="point",
            hue="color",
            linestyle='none',
            estimator=median)
plt.show()
```

Para agregar limites al final de los intervalos de confianza establezca el parametro `capsize` es igual al ancho deseado de las tapas de los intervalos de confianza.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
bins = sns.load_dataset("diamonds")
sns.catplot(x="cut",
            y="price",
            data=bins,
            kind="point",
            hue="color",
            estimator=median,
            capsize=0.2,
            errorbar=None)
plt.show()
```
