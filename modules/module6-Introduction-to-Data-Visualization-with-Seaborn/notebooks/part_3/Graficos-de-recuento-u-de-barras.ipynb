# Gráficos de recuento u de barras
Los dos primeros gráficos que veremos son los de conteo y barras.
Los gráficos de recuento y los gráficos de barras son dos tipos de visualizaciones que Seaborn llama `gráficos categoricos`.

Los gráficos categóricos implican una variable categórica que es una variable que consta de un número fijo, normalmente pequeño de posibles valores o categorías.

Este tipo de gráfico se utilizan habitualmente cuando queremos hacer comparaciones entre diferentes grupos.

`catplot()` para crear diferentes tipos de gráficos categóricos, ofreces la misma flexibilidad que `relplot()` lo hace lo que significa que sera facil crear subtramas si necesitamos usar los mismo parametros `col` y `row`.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from unicodedata import category

df = pd.read_csv('../../data/student-alcohol-consumption.csv')
sns.countplot(x='location', data=df,hue='location',kind='count')
plt.show()
```

# Cambiar el order
Podemos cambiar el orden de las barras utilizando el parámetro `order` y pasándole una lista con los valores que queremos que aparezcan en el orden que queramos.

```python
category_order = ['Rural', 'Urban']
sns.catplot(x='location',
              data=df,
              hue='location',
              order=category_order,
            kind='count')
plt.show()
```

Los graficos de barras son similares a los de conteo, pero en lugar del conteo de observaciones en cada categoria, muestra la media de una varible cuantitativa entre las observacoines de cada categoria.

```python
sns.catplot(x='sex',
            y='freetime',
            data=df,
            kind='bar',
            hue='sex')
plt.show()
```

- Seaborn muestra automaticamente intervalos de confianza del 95% para las medias.
- Muestra incertidumbre sobre nuestra estimación
- Supone que nuestros datos son una muestra aleatoria

```python
sns.catplot(y='sex',
            x='freetime',
            data=df,
            kind='bar',
            hue='sex',
            errorbar=None)
plt.show()
```
