## Tablas dinámicas y estadísticas de resumen en pandas

Las **tablas dinámicas** en pandas son simplemente DataFrames con índices ordenados. Utilizar la combinación de `.loc[]` y segmentación (slicing) resulta ideal para seleccionar subconjuntos de estas tablas de manera eficiente.

### Estadísticas de resumen en DataFrames

Los métodos para calcular estadísticas de resumen en pandas, como la media (`mean()`), admiten el argumento `axis` para especificar la dirección del cálculo:

- El valor por defecto es `axis='index'` (o `axis=0`), lo que significa que la estadística se calcula **a través de las filas** (es decir, para cada columna).
- Si quieres calcular la estadística **para cada fila**, es decir, a través de las columnas, debes fijar `axis='columns'` (o `axis=1`):

```python
bmw.mean(axis='columns')
```

Esto devuelve la media de cada fila, calculada entre sus columnas.

> **Consejo:**
> - Utiliza `axis='index'` para calcular la estadística por columna (a través de las filas).
> - Utiliza `axis='columns'` para calcular la estadística por fila (a través de las columnas).
