````python
import pandas as pd
from pyparsing import alphas

bmw = pd.read_csv('../data/BMW_Car_Sales_Classification.csv')
bmw.dtypes
````

Los graficos son una forma eficaz de compartir la informacion que has obteniendo de tus datos.
Necesitaremos importar la libreria matplotlib para poder graficar.

> import matplotlib.pyplot as plt

Vamos a graficar la columna 'Model' de nuestro dataframe bmw.

````python
import matplotlib.pyplot as plt
bmw['Model'].hist(bins = 10)
plt.show()
````

Aumentar o disminuir el numero de bins cambia la cantidad de barras que se muestran en el histograma. Lo que significa que asi podemos darnos una idea de como es la distribucion de los datos en la columna 'Model'.


El diagrama de barras puede revelar relaciones entre una variable categoria y una variable numerica.

```python
avg_bmw = bmw.groupby('Model').mean()
avg_bmw['Price'].plot(kind='bar')
```

````python
avg_bmw = bmw.groupby('Model')['Price_USD'].mean()
avg_bmw.plot(kind='bar')
plt.ylabel('Precio Promedio (USD)')
plt.title('Precio promedio por modelo BMW')
plt.ylim(0, 100000)  # Ajusta el rango según tus datos
plt.show()
````

Los graficos de lineas son estupendos para visualizar los cambios de las variables numericas a lo largo del tiempo.


````python
bmw.groupby('Year')['Price_USD'].mean().plot(x='Años',y='Precio_ USD',kind='line',rot =45)
plt.show()
````

Los diagramas de dispersión son útiles para observar la relación entre dos variables numéricas.


````python
bmw[bmw['Model'] == 'X5'].plot(x='Price_USD', y='Sales_Volume', kind='scatter')
````

Los diagramas tambien pueden superponerse.

````python
bmw[bmw['Model'] == 'X3']['Price_USD'].hist()
bmw[bmw['Model'] == 'X5']['Price_USD'].hist(alpha = 0.7)
plt.legend(['X3', 'X5'])
plt.show()
````
