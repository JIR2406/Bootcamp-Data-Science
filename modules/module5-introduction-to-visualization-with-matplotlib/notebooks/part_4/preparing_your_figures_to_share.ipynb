# Personalización de los estilos de las figuras
Si agregamos `plt.style.use('default')` al inicio de nuestro código, podemos personalizar el estilo de las figuras que generamos con Matplotlib. Esto nos permite cambiar la apariencia de las gráficas, como los colores, fuentes y otros elementos visuales.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Cargar el conjunto de datos
df = pd.read_csv('../../data/Weather Dataset.csv')
plt.style.use('ggplot')
# Crear una figura y un eje
fig, ax = plt.subplots()
ax.plot(df['Temp_C'].unique())
ax.set_xlabel('Date')
ax.set_ylabel('Pressure (kPa)')
plt.show()
```

# ¿Como elegir un estilo?
- Los fondos oscuros son ideales para presentaciones y publicaciones digitales, ya que resaltan los colores brillantes de las gráficas.
- Los fondos claros son más adecuados para documentos impresos, ya que ofrecen una mejor legibilidad y contraste con el texto.
- Los estilos como 'ggplot' o 'seaborn' son excelentes opciones para visualizaciones más complejas, ya que proporcionan una estética más pulida y profesional.
- Los estilos personalizados pueden ser útiles para mantener la coherencia con la marca o el tema de la presentación.
- Si los colores son importantes, considere usar un estilo apto para daltonicos como 'seaborn-colorblind' o 'tableau-colorblind'.

# Guardar figuras en diferentes formatos
Matplotlib permite guardar las figuras en varios formatos, como PNG, JPEG, SVG y PDF.
Solo debemos cambiar el llamado a plt.show() por plt.savefig('nombre_del_archivo.formato').

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df['Temp_C'],df['Dew Point Temp_C'],c=df.index)
ax.set_xlabel('Temp_C')
ax.set_ylabel('Dew Point Temp_C')
fig.set_size_inches(12,8)
plt.savefig('../../../../results/module5/Weather Dataset.png',dpi=300)
```
