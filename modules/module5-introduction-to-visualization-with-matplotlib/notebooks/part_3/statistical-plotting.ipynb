# Barras de error en los gráficos

Las barras de error resumen la distribución de los datos en un solo número, por ejemplo la desviación estándar de los valores.

Hay al menos dos formas diferentes de mostrar barras de error:

1. **Error absoluto**
2. **Error porcentual**

En esta ocasión, agregaremos las barras de error a un gráfico de barras. Cada llamada al método `ax.bar()` toma al menos dos argumentos:

- `x`: posiciones de las barras (por ejemplo, etiquetas o coordenadas).
- `y`: alturas de las barras (valores numéricos).

El argumento de palabra clave `yerr` recibe un valor o lista de valores numéricos que define la magnitud de la barra de error para cada punto. Adicionalmente, puedes ajustar el aspecto de las barras de error con parámetros como:

- `capsize`: longitud de la “tapa” de cada barra de error.
- `ecolor`: color de las barras de error.
- `elinewidth`: grosor de la línea de error.

A modo de ejemplo, veamos un fragmento de código en **Python / Matplotlib**:

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C']
valores    = [10, 20, 15]
errores    = [1.5, 2.0, 1.0]

fig, ax = plt.subplots()
ax.bar(
    categorias,
    valores,
    yerr=errores,       # Magnitud de la barra de error
    capsize=5,          # Tamaño de la “tapa” de la barra de error
    ecolor='red',       # Color de las líneas de error
    elinewidth=2        # Grosor de las líneas de error
)

ax.set_title('Gráfico de barras con barras de error')
ax.set_xlabel('Categorías')
ax.set_ylabel('Valores')

plt.tight_layout()
plt.show()

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../../data/Weather Dataset.csv')
fig, ax = plt.subplots()
ax.bar('Grade',
    df['Temp_C'].mean(),
    yerr=df['Temp_C'].std(),  # Error absoluto
)
ax.bar('Grade_C',
       df['Dew Point Temp_C'].mean(),
       yerr=df['Dew Point Temp_C'].std())  # Error absoluto
ax.set_ylabel('Temperature (C)')
plt.show()
````

> Es util porque resume la distribucion completa.

# Adding error bars to a line plot
Tambien podemos agregar barras de error a un gráfico de líneas. En este caso, las barras de error se aplican a los puntos de datos individuales en la línea.

````python
fig, ax = plt.subplots()
ax.errorbar(df['Date/Time'], df['Wind Speed_km/h'], yerr=df['Temp_C'].std(), label='Temp_C Error' )
ax.set_title('Line Plot with Error Bars')
ax.legend()
plt.show()
````

La segunda tecnica de visualizacion estadistica que veremos es el **boxplot**. Un boxplot es una forma de visualizar la distribución de un conjunto de datos a través de sus cuartiles. Muestra la mediana, los cuartiles y los valores atípicos.
Lo podemos llamar una secuencia de secuencias.

Este tipo de grafico nos muestra varios hitos de cada distribucion.
- La linea roja indica la altura media.
- Los bordes de la parte de la caja en el centro indican el rango intercuartil entre los perceptiles 25 y 75.
- Los bigotes de los extremos de las barras delgadas indicar una vez y media el tamaño del rango intercuartil mas alla de los perceptiles 75 y 25.
- Esto deberia abarcar aproximadamente el 99 por ciento de la distribucion si los datos son gaussianos o normales.
- Los puntos fuera de los bigotes son considerados valores atipicos.

````python
fig, ax = plt.subplots()
ax.boxplot([df['Temp_C'],df['Dew Point Temp_C']])
ax.set_xticklabels(['Temp_C', 'Dew Point Temp_C'])
ax.set_ylabel('Temperature (C)')
plt.show()
````
