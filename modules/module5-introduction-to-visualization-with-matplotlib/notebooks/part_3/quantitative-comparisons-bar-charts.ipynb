El siguiente grafico de barras muestra la cantidad de canciones escuchadas por plataforma.

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../../data/spotify_history.csv')
fig, ax = plt.subplots()
ax.bar(df['platform'],df.index)
plt.show()
````

## Interlude: rotate the tick labels
Podemos arreglar las etiquetas, se pueden rotar 45 grados para que no se vean tan mal. Que tenga una mejor apariencia.

```python
ax.set_xticklabels(df['platform'], rotation=45)
ax.set_ylabel('Cantidad de canciones escuchadas')
```

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../../data/spotify_history.csv')
fig, ax = plt.subplots()
ax.bar(df['platform'], df.index)
plt.xticks(rotation=45)  # Esto rota las etiquetas automáticamente
ax.set_ylabel('Cantidad de canciones escuchadas')
plt.show()
````

Podemos agregar el numero de medallas de bronce estableciendo el limite inferior de las barras, que sera la suma del numero de medallas de oro y el numero de medallas de plata.

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])

ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"])
ax.bar(medals.index, medals["Bronze"],
       bottom=medals["Gold"] + medals["Silver"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
plt.show()
```

Para lograr que sea mas facil de leer podemos agregar `legend()` para que se muestre la leyenda de cada barra.
```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"], label="Gold")
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"], label="Silver")
ax.bar(medals.index, medals["Bronze"],
       bottom=medals["Gold"] + medals["Silver"],
       label="Bronze")

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
ax.legend()
plt.show()
```
