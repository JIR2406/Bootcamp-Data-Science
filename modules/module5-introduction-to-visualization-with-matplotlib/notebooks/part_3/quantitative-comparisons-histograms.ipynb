> Un histograma mostraria la distrubucion completa de valores dentro de cada variable.

# Histogramas: altura de las barras y configuración de bins

- La altura de cada barra representa el número de observaciones dentro de un intervalo específico de valores (también llamado contenedor, clase o bin).
- De manera predeterminada, un histograma utiliza 10 contenedores (bins), pero puedes personalizarlos.

## Cómo configurar `bins`
- Entero: indica el número total de contenedores que deseas.
- Secuencia de valores: se interpreta como los límites de los contenedores. Deben ser valores numéricos ordenados que definan los bordes entre intervalos.

Ejemplo: `bins = [0, 5, 10, 20]` crea tres contenedores: [0, 5), [5, 10) y [10, 20].

````python
fix,ax = plt.subplots()
ax.hist(df['Dew Point Temp_C'])
ax.hist(df['Temp_C'])
ax.set_xlabel('Temperature (C)')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Temperature Variables')
ax.legend(['Dew Point Temp_C', 'Temp_C'])
plt.show()
````

# Personalización de Histogramas

En ocasiones, queremos **eliminar la oclusión** (superposición) entre las barras de un histograma.
Por defecto, muchos librerías (por ejemplo, Matplotlib) usan un histograma de tipo `bar`, que rellena el área de cada barra.
Si preferimos ver sólo los contornos, podemos pasar el tipo `step`.

---

## 1. Histograma por defecto (bar)

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data,
         bins=5,
         histtype='bar',   # Tipo por defecto: rellena cada barra
         color='skyblue',
         edgecolor='black')

plt.title('Histograma tipo "bar"')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

---

## 2. Histograma sin relleno (step)

```python
import matplotlib.pyplot as plt

# Mismos datos
plt.hist(data,
         bins=5,
         histtype='step',  # Sólo contorno, sin relleno
         color='crimson',
         linewidth=1.5)

plt.title('Histograma tipo "step"')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

---

## 3. Resumen de opciones

- **histtype='bar'**
  Rellena cada barra. Útil cuando no hay solapamiento o queremos ver el área acumulada.

- **histtype='step'**
  Dibuja sólo el contorno. Perfecto para comparar varias distribuciones sin oclusión.

- **bins**
  Número de compartimentos (barras) en el histograma.

- **color / edgecolor / linewidth**
  Personaliza colores, borde y grosor de las líneas.

---

````python
fix,ax = plt.subplots()
ax.hist(df['Dew Point Temp_C'],histtype='step')
ax.hist(df['Temp_C'],histtype='step')
ax.set_xlabel('Temperature (C)')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Temperature Variables')
ax.legend(['Dew Point Temp_C', 'Temp_C'])
plt.show()
````
