> ¿Que pasa si desea comparar los valores de diferentes variables a lo largo de las observaciones?

Esto a veces se denomina *comparación bivariada*, porque involucra los valores de dos variables diferentes.
Una visualizacion estandar para comparaciones bivariadas es un diagrama de dispersion.

```python
import pandas as pd
df = pd.read_csv('../../data/Weather Dataset.csv')
print("Importacion de datos exitosa")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df['Temp_C'],df['Dew Point Temp_C'],c=df.index)
ax.set_xlabel('Temp_C')
ax.set_ylabel('Dew Point Temp_C')
plt.show()
```

Podemos personalizar los graficos de dispersion de una manera similar a la personalizacion que introdujimos en otros graficos.
