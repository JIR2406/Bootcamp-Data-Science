Las series de tiempo y las visualizaciones de series de tiempo son una excelente herramienta para detectar patrones en los datos.
Las variables continuas como las precipitaciones o las temperaturas son organizados en nuestra tabla de datos segun una variable temporal, los meses del año.


Si queremos que pandas detecte una serie de tiempo necesitaremos indicarle que analice la columna `fecha` como una fecha.



```python
import pandas as pd
from matplotlib import pyplot as plt

spoti = pd.read_csv('../../data/spotify_history.csv')
spoti['ts'] = pd.to_datetime(spoti['ts']) # Convertimos la columna 'ts' a tipo datetime
spoti.dtypes
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(spoti['ts'])
ax.set_xlabel('Cantidad de reproducciones')
ax.set_ylabel('Año')
plt.show()
```
