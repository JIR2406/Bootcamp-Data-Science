## Small Multiples

A veces, agregar demasiados datos a un solo gráfico puede sobrecargarlo, haciendo que los patrones importantes pasen desapercibidos.
Para abordar este problema, podemos utilizar **small multiples** o **múltiplos pequeños**: varios gráficos pequeños que muestran datos similares bajo diferentes condiciones o categorías.

En Matplotlib, los small multiples se implementan mediante **subplots** (subparcelas). De hecho, la función que nos permite crear estos gráficos pequeños se llama justamente `subplots`.

Los múltiples gráficos suelen organizarse en una cuadrícula de filas y columnas dentro de la página o figura, permitiendo comparar fácilmente los patrones entre diferentes conjuntos de datos.

Por ejemplo, a continuación se muestra cómo crear una figura con tres filas y dos columnas de subparcelas en Matplotlib, ideal para visualizar comparaciones paralelas de manera clara y ordenada.

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3,2)
plt.show()
````

> En cambio, se trata de una matriz de objetos `Axes` con una forma de 3 x 2

````python
ax.shape
````

Para agregar datos, ahora tendriamos que indexar en este objeto y llamar al metodo plot un elemento de la matriz.

Existe un caso especial para aquellas situaciones en las que solo se tiene una fila o una columna de graficos.

```python
fix,ax = plt.subplot(2,1)
```

> En este caso la matriz resultante sera unidimensional, y solo tendra que proporcionar un indice para acceder a los elementos de esta matriz.

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"], linestyle='--', color='b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"], linestyle='--', color='b')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"], linestyle='--', color='r')

ax[0].set_ylabel("Precipitation (inches)")

ax[1].set_xlabel("Time (months)")
```

Una cosa que todavia debemos tener en cuenta es el rango del eje y en los dos graficos, que no es exactamente el mismo.

Para asegurarnos que todas las subparcelas tengan el mismo rango de valores del eje y, podemos usar el metodo `sharey=True` al crear las subparcelas.

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, sharey=True)
seatle = [70, 65, 60, 55, 50, 45, 40, 45, 50, 55, 60, 65]
austin = [50, 55, 60, 65, 70, 75, 80, 85, 80, 75, 70, 65]

ax[0].plot(range(1, 13), seatle, color='b')
ax[1].plot(range(1, 13), austin, color='r')
ax[0].set_ylabel("Temperature (F)")
ax[0].set_title("Seattle")
ax[1].set_title("Austin")
ax[1].set_ylabel("Temperature (F)")
ax[1].set_xlabel("Time (months)")

plt.subplots_adjust(hspace=0.4)  # Ajusta el espacio vertical entre subplots
plt.show()
````
