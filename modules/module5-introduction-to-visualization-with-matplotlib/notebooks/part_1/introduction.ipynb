# Introducción de visualización de datos con Matplotlib
Las visualizaciones de datos le permiten obtener información de los datos y comunicarse sobre ellos con otros.
Hay muchas bibliotecas de software que visualizan datos.

Una de las principales ventajas de Matplotlib es que te da control total sobre las propiedades de tu grafico.
Esto le permite personalizar y controlar las propiedades precisas de sus visualizaciones.

Existen muchas maneras diferentes de utilizar Matplotlib.
- Utilizaremos la interfaz principal orientada a objetos.
- Esta interfaz se proporciona a traves del submodulo `pyplot`.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() # Cuando se llama sin ninguna entrada, crea dos objetos: una figura y un eje.
plt.show() # Muestra la figura en una ventana emergente.
```

> Ahora añadiremos datos a la figura.

```python
import pandas as pd
spoti = pd.read_csv('../../data/spotify_history.csv')
fig, ax = plt.subplots()
ax.plot(spoti['platform'].unique(),[1,5,4,6,7,9]) # Añade datos al eje.
ax.plot(spoti['platform'].unique(),[10,12,20,25,32,26]) # Añade datos al eje.
plt.show()
```
