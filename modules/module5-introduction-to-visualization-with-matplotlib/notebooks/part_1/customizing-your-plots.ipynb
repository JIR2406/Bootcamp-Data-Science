## Personalización de la apariencia de los datos en el gráfico

Lo primero que vamos a personalizar es la **apariencia de los datos** en el gráfico.
El método de trazado (`plot`) acepta un argumento de palabra clave opcional llamado `marker`, que permite indicar el interés en agregar **marcadores** a la trama y también especificar el **tipo de marcador** que se desea utilizar.


````python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import lineStyles

spoti = pd.read_csv('../../data/spotify_history.csv')
fig, ax = plt.subplots() # Cuando se llama sin ninguna entrada, crea dos objetos: una figura y un eje.
ax.plot(spoti['platform'].unique(),[1,5,4,6,7,9],marker = 'o') # Añade datos al eje.
plt.show() # Muestra la figura en una ventana emergente.
````

> Al pasar la letra minuscula `o` indica que desea utilizar circulos como marcadores.

En el ejemplo anterior, el argumento `marker='o'` agrega **círculos** en cada punto de datos del gráfico.

### Tipos de marcadores comunes

| Código | Descripción      |
|--------|------------------|
| `'o'`  | Círculo          |
| `'s'`  | Cuadrado         |
| `'^'`  | Triángulo arriba |
| `'D'`  | Diamante         |
| `'x'`  | X                |
| `'+'`  | Más              |

Tambien se puede ir aun mas alla y enfatizar esto cambiando la apariencia de las lineas de conexion.

Para eso utilizamos el argumento `linestyle`
Podemos darle `--` para que las lineas sean discontinuas, o `-` para que sean continuas.

Inclusive podriamos llegar a eliminarlo pasando la cadena `None` como argumento.

````python
fig, ax = plt.subplots() # Cuando se llama sin ninguna entrada, crea dos objetos: una figura y un eje.
ax.plot(spoti['platform'].unique(),[1,5,4,6,7,9],marker = 'o',linestyle='--') # Añade datos al eje.
plt.show() # Muestra la figura en una ventana emergente.
````

Es posible cambiar el color de los marcadores y las líneas de conexión utilizando el argumento `color`.
Podemos pasarle un nombre de color, un código hexadecimal o un valor RGB.

````python
fig, ax = plt.subplots() # Cuando se llama sin ninguna entrada, crea dos objetos: una figura y un eje.
ax.plot(spoti['platform'].unique(),[1,5,4,6,7,9],marker = 'o',linestyle='--',color='r') # Añade datos al eje.
plt.show() # Muestra la figura en una ventana emergente.
````
