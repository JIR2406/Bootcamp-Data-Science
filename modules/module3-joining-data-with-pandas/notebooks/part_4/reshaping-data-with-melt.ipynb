En este notebook hablaremos acerca del metodo 'melt' de pandas, el cual nos permite transformar un DataFrame de formato ancho a formato largo.

En general loa datos con formato ancho son mas faciles de leer qie los con formato largo.

El metodo de fusion nos permite despivotar o cambiar el formato de nuestro conjunto de dato.

```python
import pandas as pd

movies = pd.read_csv('../../data/TMDB_movie_dataset_v11.csv')

movies.head(3)
```

```python
# El metodo melt nos permite transformar un DataFrame de formato ancho a formato largo.
# En este caso, vamos a transformar el DataFrame de peliculas para que cada fila sea una
# pelicula y cada columna sea un atributo de la pelicula.
movies_melted = movies.melt(id_vars=['id', 'title'])
movies_melted.head(10)
```

```python

# Selecciona las columnas a derretir (todas excepto id, title y release_date)
value_vars = [col for col in movies.columns if col not in ['id', 'title', 'release_date']]

movies_melted = movies.melt(id_vars=['id', 'title', 'release_date'], value_vars=value_vars, var_name='variable', value_name='valor')
movies_melted.head(10)
```
