# Análisis del método `query()` de Pandas para seleccionar datos

En este notebook, analizaremos el método `query()` de Pandas, una de las formas más eficientes y legibles para seleccionar datos de un DataFrame.

Pandas proporciona múltiples métodos para filtrar y seleccionar información; `query()` es uno de los más versátiles, permitiendo realizar consultas complejas de manera sencilla utilizando expresiones similares a SQL.

## ¿Cómo funciona el método `query()`?

El método `query()` recibe como argumento una cadena de texto con la condición que se utilizará para seleccionar las filas del DataFrame.

**Sintaxis básica:**
```python
DataFrame.query('CONDICIÓN_DE_SELECCIÓN')
```

La condición que se expresa en el método `query()` es muy similar a la que se usaría después de la cláusula `WHERE` en una sentencia SQL.

> Por ejemplo:
> ```python
> df.query('columna > 5 and columna2 == "valor"')
> ```

Esto permite escribir filtros complejos de manera más intuitiva y legible, facilitando el análisis y manipulación de los datos.

````python
import pandas as pd

movies = pd.read_csv('../../data/TMDB_movie_dataset_v11.csv')

movies.query('vote_average == 8.364')
````
